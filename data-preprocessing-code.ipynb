```python
import os
import cv2
from tqdm import tqdm
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Change this according to your dataset location

DATASET_PATH = r"Dataset"

TRAIN_PATH = os.path.join(DATASET_PATH, "Train")
VAL_PATH = os.path.join(DATASET_PATH, "Validation")
TEST_PATH = os.path.join(DATASET_PATH, "Test")
```

```python
for folder in [TRAIN_PATH, VAL_PATH, TEST_PATH]:
    print("\n", folder)
    print(os.listdir(folder))
```

```python
def count_images(folder):

    total = 0

    for cls in os.listdir(folder):

        class_path = os.path.join(folder, cls)

        if os.path.isdir(class_path):

            count = len(os.listdir(class_path))

            print(f"{cls}: {count}")

            total += count

    print("Total:", total)
```

```python
print("Training Set")
count_images(TRAIN_PATH)

print("\nValidation Set")
count_images(VAL_PATH)

print("\nTest Set")
count_images(TEST_PATH)
```

```python
def check_image_sizes(folder, num_images=5):

    for cls in os.listdir(folder):

        class_path = os.path.join(folder, cls)

        print(f"\nClass: {cls}")

        images = os.listdir(class_path)[:num_images]

        for img in images:

            path = os.path.join(class_path, img)

            image = Image.open(path)

            print(img, image.size)
```

```python
check_image_sizes(TRAIN_PATH)
```

```python
def resize_images(input_folder, output_folder, size=(224, 224)):

    os.makedirs(output_folder, exist_ok=True)

    for cls in os.listdir(input_folder):

        input_class = os.path.join(input_folder, cls)

        output_class = os.path.join(output_folder, cls)

        os.makedirs(output_class, exist_ok=True)

        images = os.listdir(input_class)

        for image_name in tqdm(images, desc=cls):

            input_path = os.path.join(input_class, image_name)

            output_path = os.path.join(output_class, image_name)

            image = cv2.imread(input_path)

            if image is None:
                continue

            image = cv2.resize(image, size)

            cv2.imwrite(output_path, image)
```

```python
OUTPUT_PATH = r"Dataset-Preprocessed"

TRAIN_OUTPUT = os.path.join(OUTPUT_PATH, "Train")
VAL_OUTPUT = os.path.join(OUTPUT_PATH, "Validation")
TEST_OUTPUT = os.path.join(OUTPUT_PATH, "Test")
```

```python
resize_images(TRAIN_PATH, TRAIN_OUTPUT)

resize_images(VAL_PATH, VAL_OUTPUT)

resize_images(TEST_PATH, TEST_OUTPUT)
```
